```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Calculate factorial of 12
result = factorial(12)
print("Factorial of 12 is:", result)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics

# Define true positive rates (TPR) and false positive rates (FPR)
tpr = [0.18, 0.55, 0.89, 1.00]
fpr = [0.10, 0.30, 0.50, 0.70]

# Calculate AUC
auc = metrics.auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()

print(f'AUC: {auc:.2f}')
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# Example data
data = {'value': [10, 12, 14, 15, 18, 19, 20, 22, 25, 100, 10, 12, 14, 15, 18, 19, 20, 22, 25, 100, 10, 12, 14, 15, 18, 19, 20, 22, 25, 100, 10, 12, 14, 15, 18, 19, 20, 22, 25, 100, 10, 12, 14, 15, 18, 19, 20, 22, 25, 100, 10, 12, 14, 15, 18, 19, 20, 22, 25, 100]}
df = pd.DataFrame(data)

# Calculate the z-scores for each value
z_scores = np.abs(stats.zscore(df['value']))

# Set a threshold for the z-score
threshold = 3

# Filter out the outliers
df_filtered = df[(z_scores < threshold)]

print("Data after removing outliers:")
print(df_filtered)
```

```python
import numpy as np
import pandas as pd

# Example data with missing values
data = {'value': [10, 12, 14, np.nan, 18, 19, 20, 22, 25, 100]}
df = pd.DataFrame(data)

# Mean Imputation
mean_value = df['value'].mean()
df['value'].fillna(mean_value, inplace=True)
print("Data after mean imputation:")
print(df)
# Median Imputation
median_value = df['value'].median()
df['value'].fillna(median_value, inplace=True)
print("Data after median imputation:")
print(df)

# Mode Imputation
mode_value = df['value'].mode()[0]
df['value'].fillna(mode_value, inplace=True)
print("Data after mode imputation:")
print(df)

# IQR Calculation for upper and lower tail imputation
Q1 = df['value'].quantile(25)
Q3 = df['value'].quantile(75)
IQR = Q3 - Q1

# Upper Tail Imputation
upper_tail = Q3 + 1.5 * IQR
df['value'] = df['value'].apply(lambda x: upper_tail if x > upper_tail else x)
print("Data after upper tail imputation:")
print(df)

# Lower Tail Imputation
lower_tail = Q1 - 1.5 * IQR
df['value'] = df['value'].apply(lambda x: lower_tail if x < lower_tail else x)
print("Data after lower tail imputation:")
print(df)
```
